```python
from openai import OpenAI
from IPython.display import Markdown

client = OpenAI()
```

## MCP Fetch

```python
# Call the OpenAI API with responses endpoint
response1=client.responses.create(
    model="gpt-4.1",
    input="Find information about the Restaurant Paradise in Hyderabad",
    tools=[
        {
            "type":"mcp",
            "server_label":"fetch",
            "server_url":"https://remote.mcpservers.org/fetch/mcp",
            "require_approval":"never"

            }
    ]
)
```

```python
Markdown(response1.output_text)
```

```python
response1.output[0]
```

```python
response1.tools
```
